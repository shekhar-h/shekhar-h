### Segment 5 - Parametric Correlation Analysis

# Correlation does not imply causation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pylab import rcParams

import seaborn as sb
import scipy
from scipy.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
plt.style.use('seaborn-whitegrid')
```

#### The Pearson Coefficient

```python
address = 'C:/Users/Vibhanshu/Documents/GitHub/Python For Data Science/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.rename(columns = {'Unnamed: 0':'car_names'}, inplace = True)
cars.index = cars['car_names']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['mpg','hp','qsec','wt']]
sb.pairplot(x)
```

#### Using scipy to calculate the Pearson correlation coefficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']

pearsonr_coeff, p_value = pearsonr(mpg,hp)
print('PearsonR Correlation Coeff %0.3f'% (pearsonr_coeff))
print('p-value %0.10f'% (p_value))
```

```python
pearsonr_coeff, p_value = pearsonr(mpg,qsec)
print('PearsonR Correlation Coeff %0.3f'% (pearsonr_coeff))
print('p-value %0.10f'% (p_value))
```

```python
pearsonr_coeff, p_value = pearsonr(mpg,wt)
print('PearsonR Correlation Coeff %0.3f'% (pearsonr_coeff))
print('p-value %0.10f'% (p_value))
```

### Segment 6 - Nonparametric correlation analysis

#### Segment 6 - Delving into non-parametric methods using pandas and scipy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
import scipy
from scipy.stats import spearmanr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

#### The Spearman Rank Correlation

```python
address = 'C:/Users/Vibhanshu/Documents/GitHub/Python For Data Science/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.rename(columns = {'Unnamed: 0':'car_names'}, inplace = True)
cars.index = cars['car_names']
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['cyl','vs','am','gear']]
sb.pairplot(x)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coeff, p_value = spearmanr(cyl,vs)
print('Spearman Rank correlation coefficient %0.3f' %(spearmanr_coeff))
print('p-Value %0.10f' %(p_value))
```

```python
spearmanr_coeff, p_value = spearmanr(cyl,am)
print('Spearman Rank correlation coefficient %0.3f' %(spearmanr_coeff))
print('p-Value %0.10f' %(p_value))
```

```python
spearmanr_coeff, p_value = spearmanr(cyl,gear)
print('Spearman Rank correlation coefficient %0.3f' %(spearmanr_coeff))
print('p-Value %0.10f' %(p_value))
```

#### Chi-square test for independence

```python
table = pd.crosstab(cyl,am)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-Square statistic %0.3f p-value %0.5f' %(chi2, p))
```

```python
table = pd.crosstab(cyl, vs)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-Square statistic %0.3f p-value %0.5f' %(chi2, p))
```

```python
table = pd.crosstab(cyl, gear)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-Square statistic %0.3f p-value %0.5f' %(chi2, p))
```

#### Segment 7 - Transforming Dataset Transformations

```python
import numpy as np
import pandas as pd
import scipy

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb

import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

#### Normalizing and transforming features with MinMaxScalar() and fit_transform()

```python
address = 'C:/Users/Vibhanshu/Documents/GitHub/Python For Data Science/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.rename(columns = {'Unnamed: 0':'car_names'}, inplace = True)
cars.index = cars['car_names']
cars.head()
```

```python
mpg = cars.mpg

plt.plot(mpg)
```

```python
cars[['mpg']].describe()
```

```python
# reshaping the mpg variable into 1-column matrix
mpg_matrix = mpg.values.reshape(-1,1)

# MinMaxScalar() scales the function we use to transform the variable by sclaing it to a defined range; by default the above range is (0,1)
# fit_transform() to fit the minmaxscalar transformer to the data and return a transformed version of the data
scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
# to define a custom range b/w 0 and 10
scaled = preprocessing.MinMaxScaler(feature_range = (0,10))

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

#### Using scale() to scale the features

```python
# To UN-scale and UN-transform the variables
standardized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)
plt.plot(standardized_mpg)
```

```python
# To UN-scale and UN-transform the variables without mean/std parameters
standardized_mpg = scale(mpg, axis=0)
plt.plot(standardized_mpg)
```
