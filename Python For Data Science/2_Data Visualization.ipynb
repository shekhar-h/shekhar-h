### Segement 2 - Defining Elements of a plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

Two methods for Plot Building:<br>
    1. Functional method - Discussed above<br>
    2. Object oriented method
    
    
    Steps in Object-oriented Plotting
    i) Create a blank figure object        
    ii) Add axes to the figure
    iii) Generate plot(s) within figure object
    iv) Specify plotting and layout parameters for the plots within the figure
        

```python
%matplotlib inline 
# Tells python to print the plots inline
rcParams['figure.figsize']=5,4
```

#### Defining axes, ticks and grinds

```python
x = range(1,10)
y = [1,3,2,5,0,3,8,2,4]

fig = plt.figure() # prints a blank figure
# add_axes([a,b,c,d]) - a,b is the starting point at southwest corner of the figure, c=width, d=height
ax = fig.add_axes([0.1,0.1,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure() 
ax = fig.add_axes([0.1,0.1,1,1])

ax.set_xlim([0,9])
ax.set_ylim([0,9])

ax.set_xticks([0,1,2,4,5,6,8,9])
ax.set_yticks([2,4,6,8,10])

ax.plot(x,y)
```

```python
fig = plt.figure() 
ax = fig.add_axes([0.1,0.1,1,1])

ax.set_xlim([0,9])
ax.set_ylim([0,9])

ax.grid() # adds grid for increased readability

ax.plot(x,y)
```

#### Generating multiple plots in one figure with subplots

```python
fig = plt.figure()
# ax1,ax2 - in a tuple put the names of however may subplots are required - here using 2
# in the plt.subplots() - the parameters define how many rows and columns are needed to show those subplots - 1 row and 2 cols
fig,(ax1,ax2) = plt.subplots(1,2) 

ax1.plot(x)
ax2.plot(x,y)
```

#### The functional Method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
# adding x and y axis labels
plt.bar(x,y)
plt.xlabel('Your X-axis Label')
plt.ylabel('Your y-axis label')
```

```python
z = [1,2,3,4,.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']
# adding value labels
plt.pie(z, labels=veh_type)
plt.show()
```

#### The object-oriented Method

```python
address = 'C:/Users/Vibhanshu/Documents/GitHub/Python For Data Science/Exercise Files/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.rename(columns = {'Unnamed: 0':'car_names'}, inplace = True)
cars.head()
```

```python
mpg = cars['mpg']

fig = plt.figure()
ax= fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 60, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in MT Cars Dataset')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')
```

#### Adding a legend to the plot

#### The functional Method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented Method

```python
mpg = cars['mpg']

fig = plt.figure()
ax= fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation = 90, fontsize='medium')
ax.set_title('Miles Per Gallon of Cars in MT Cars Dataset')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')

ax.legend(loc='best')
```
